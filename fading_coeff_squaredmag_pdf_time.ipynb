```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a random fading channel
def generate_random_h(K):
    return np.random.normal(0, np.sqrt(0.5), (K, 1)) + 1j * np.random.normal(0, np.sqrt(0.5), (K, 1))

# Function to add L independent random channels
def generate_and_sum_random_h(L, K):
    # Initialize the sum to zero
    combined_mag = np.zeros((K, 1))

    # Loop L times, each time generating a new random h and adding its squared magnitude
    for _ in range(L):
        h = generate_random_h(K)
        combined_mag += (np.abs(h))**2

    return combined_mag

# Number of samples per channel
K = 100000
L_max = 10  # Set the maximum number of independent channels (L)

# Create a figure for plotting
plt.figure(figsize=(10, 6))

# Loop over different values of L (from 1 to L_max)
for L in range(1, L_max + 1):
    # Generate L independent channels and sum their squared magnitudes
    combined_mag = generate_and_sum_random_h(L, K)

    # Plot the histogram for the current L
    plt.hist(combined_mag, bins=1000, alpha=0.7, label=f"L={L}", density=True)

# Finalizing the plot
plt.xlabel('Squared Magnitude')
plt.ylabel('Density')
plt.title('Histogram of Squared Magnitudes for Time Diversity (L=1 to L=10)')
plt.legend(loc='upper right', bbox_to_anchor=(1.05, 1), fontsize=8)
plt.grid(True)
plt.show()
```
